```python
#import the libraries
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
#input two paragraphs
text = """Natural Language Processing (NLP) is a field of Artificial Intelligence. 
         It helps computers understand, interpret and generate human language. 
         Many applications, such as: chatbots, translation systems, and voice assistants use NLP techniques.

         Data preprocessing is an important step in NLP. Raw text often contains punctuation marks, 
         stop words, and unnecessary symbols. Cleaning the text improves the performance of machine learning models; 
         therefore, preprocessing is essential."""
```

```python
words = word_tokenize(text)
print("Tokenized Words:\n", words)
```

```python
#Removing Punctuation Marks
words_no_punc = [word for word in words if word not in string.punctuation]
print("\nAfter Removing Punctuation:\n", words_no_punc)
```

```python
#Stop Word Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_no_punc if word.lower() not in stop_words]
print("\nAfter Stopword Removal:\n", filtered_words)
```

```python
#Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("\nAfter Stemming:\n", stemmed_words)
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nAfter Lemmatization:\n", lemmatized_words)
```
